```python
plt.figure(figsize=(16,10))
fig,axis=plt.subplots()
sb.heatmap(scaled.corr(), mask=mask, annot=True, annot_kws={"size": 10}, fmt=".2f", cmap='RdBu_r', vmin=-1, vmax=1)
plt.xticks(fontsize=12, rotation=60)
plt.yticks(fontsize=12, rotation=0)
plt.title("Correlation Coefficients (standardized values)")
axis.set_xticklabels(['Total Market', 'CRIX', 'F5', 'Bitwise10', 'Bitwise20', 'Bitwise70', 'Bitwise100', 'CCI30'])
axis.set_yticklabels(['Total Market', 'CRIX', 'F5', 'Bitwise10', 'Bitwise20', 'Bitwise70', 'Bitwise100', 'CCI30'])
plt.show()
```

```python
indices=indices.drop(['Total_Market_s', 'CRIX_s', 'F5_s', 'Bitwise10_s', 'Bitwise20_s', 'Bitwise70_s', 'Bitwise100_s', 'CCI30_s'], axis=1)
```

```python
# Computation of log returns
indices['Total_Market_log_return'] = np.log(indices['Total_Market']/indices['Total_Market'].shift(1))
indices['CRIX_log_return'] = np.log(indices['CRIX']/indices['CRIX'].shift(1))
indices['F5_log_return'] = np.log(indices['F5']/indices['F5'].shift(1))
indices['Bitwise10_log_return'] = np.log(indices['Bitwise10']/indices['Bitwise10'].shift(1))
indices['Bitwise20_log_return'] = np.log(indices['Bitwise20']/indices['Bitwise20'].shift(1))
indices['Bitwise70_log_return'] = np.log(indices['Bitwise70']/indices['Bitwise70'].shift(1))
indices['Bitwise100_log_return'] = np.log(indices['Bitwise100']/indices['Bitwise100'].shift(1))
indices['CCI30_log_return'] = np.log(indices['CCI30']/indices['CCI30'].shift(1))
indices.head()
```

```python
indices.round(3)
```

```python
# Plotting log returns
plt.plot(indices.Date, indices.Total_Market_log_return)
plt.title("Log Returns")
plt.xlabel("Year")
plt.ylabel("returns")
plt.legend(["Total Market"], loc= (1.1,0.5))
plt.show()
```

```python
plt.plot(indices.Date, indices.CRIX_log_return)
plt.title("Log Returns")
plt.xlabel("Year")
plt.ylabel("returns")
plt.legend(["CRIX"], loc= (1.1,0.5))
plt.show()
```

```python
plt.plot(indices.Date, indices.Total_Market_log_return)
plt.plot(indices.Date, indices.CRIX_log_return)
plt.title("Log Returns")
plt.xlabel("Year")
plt.ylabel("returns")
plt.legend(["Total Market", "CRIX"], loc= (1.1,0.5))
plt.show()
```
